### 1. 行业薪资水平分布

分析“钱”景是数据挖掘中最直观的部分。

- **薪资直方图与核密度估计 (KDE):** 统计 `平均薪资(k)` 的分布情况，观察薪资主要集中在哪个区间（比如 10k-15k 还是 20k-30k）。

- **薪资箱线图 (Boxplot):** 按 `学历要求` 或 `公司性质` 分组，查看不同群体的薪资中位数、最大/最小值以及离群值（高薪岗位）。

- **分析价值：** 揭示学历是否真的能带来显著的溢价。
